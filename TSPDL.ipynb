```python
#BUSQUEDA LOCAL

import time
inicio = time.time()

# Código a medir
time.sleep(1)
import random
import itertools

def calculate_distance(locations, path):
    distance = 0
    for i in range(len(path) - 1):
        start = path[i]
        end = path[i + 1]
        distance += locations[start][end]
    return distance

def check_draft_limits(path, depths, draft_limits):
    max_depth = max(depths)
    for port in path:
        if depths[port] > max_depth or draft_limits[port] > depths[port]:
            return False
    return True

def get_port_index(port_names, port_name):
    for i in range(len(port_names)):
        if port_names[i] == port_name:
            return i
    return -1

def get_nearest_neighbor(locations, path, current_port):
    min_distance = float('inf')
    nearest_port = None
    for port in path:
        if locations[current_port][port] < min_distance:
            min_distance = locations[current_port][port]
            nearest_port = port
    return nearest_port

def local_search(locations, port_names, depths, draft_limits, start_port):
    num_ports = len(locations)
    current_port = get_port_index(port_names, start_port)
    unvisited_ports = list(range(num_ports))
    unvisited_ports.remove(current_port)
    path = [current_port]

    while unvisited_ports:
        nearest_port = get_nearest_neighbor(locations, unvisited_ports, current_port)
        unvisited_ports.remove(nearest_port)
        if check_draft_limits(path + [nearest_port], depths, draft_limits):
            path.append(nearest_port)
            current_port = nearest_port
        else:
            break

    path.append(get_port_index(port_names, start_port))

    return path

# Generar ubicaciones al azar
num_ports = 10000  # Cambiar el número de puertos aquí (10, 20, 30)
port_names = [f"Puerto {i}" for i in range(1, num_ports + 1)]
locations = [[0] * num_ports for _ in range(num_ports)]
for i in range(num_ports):
    for j in range(num_ports):
        if i != j:
            # Simulación de distancias al azar
            locations[i][j] = random.randint(1, 100)

profundidades_de_puertos = [random.uniform(5.0, 15.0) for _ in range(num_ports)]

start_port = random.choice(port_names)  # Puerto de inicio aleatorio

start_index = get_port_index(port_names, start_port)
if start_index != -1:
    path = local_search(locations, port_names, profundidades_de_puertos, profundidades_de_puertos, start_port)
    total_distance = calculate_distance(locations, path)

    print(f"\nLa ruta más eficiente que cumple con los límites de calado y comienza en {start_port} es:")

    for i in range(len(path)):
        port = path[i]
        if i == 0:
            print(f"\nPuerto: {port_names[port]}")
        elif i == len(path) - 1:
            print(f"Puerto de regreso: {port_names[port]}")
        else:
            print(f"Puerto intermedio: {port_names[port]}")

    print(f"Distancia total recorrida: {total_distance} unidades de distancia")
else:
    print("El puerto de inicio no se encontró en la lista de puertos.")
# -------------

fin = time.time()
print(fin-inicio)
import resource

# Obtener el consumo de memoria actual en bytes
memory_usage = resource.getrusage(resource.RUSAGE_SELF).ru_maxrss
print(f"Consumo de memoria: {memory_usage} bytes")
```

```python
import itertools
import time
inicio = time.time()

# Código a medir
time.sleep(1)

def calculate_distance(locations, path):
    distance = 0
    for i in range(len(path) - 1):
        start = path[i]
        end = path[i + 1]
        distance += locations[start][end]
    return distance

def get_port_index(port_names, port_name):
    for i in range(len(port_names)):
        if port_names[i] == port_name:
            return i
    return -1

def get_nearest_neighbor(locations, unvisited_ports, current_port):
    min_distance = float('inf')
    nearest_port = None
    for port in unvisited_ports:
        if locations[current_port][port] < min_distance:
            min_distance = locations[current_port][port]
            nearest_port = port
    return nearest_port

def nearest_neighbor(locations, port_names, start_port):
    num_ports = len(locations)
    current_port = get_port_index(port_names, start_port)
    unvisited_ports = list(range(num_ports))
    unvisited_ports.remove(current_port)
    path = [current_port]

    while unvisited_ports:
        nearest_port = get_nearest_neighbor(locations, unvisited_ports, current_port)
        unvisited_ports.remove(nearest_port)
        path.append(nearest_port)
        current_port = nearest_port

    path.append(get_port_index(port_names, start_port))

    return path

# Función para calcular el calado del barco
def calcular_calado(longitud_barco, ancho_barco, altura_casco, peso_carga, densidad_agua):
    # Calcula el volumen del casco
    volumen_casco = longitud_barco * ancho_barco * altura_casco

    # Calcula el volumen de la carga
    volumen_carga = peso_carga / densidad_agua

    # Calcula el volumen total sumergido
    volumen_total = volumen_casco + volumen_carga

    # Calcula el área de la sección transversal del casco
    area_casco = longitud_barco * ancho_barco

    # Calcula el calado
    calado = volumen_total / area_casco

    return calado

# Datos de entrada
locations = [
    [0, 90, 402.03, 452.47, 404.75],
    [90, 0, 477.95, 514.14, 476.68],
    [402.03, 477.95, 0, 56.26, 53.72],
    [452.47, 514.14, 56.26, 0, 98.53],
    [404.75, 476.68, 53.72, 98.53, 0]
]
port_names = ["Puerto marítimo de San Antonio", "Puerto marítimo de Valparaiso", "Puerto marítimo de San Vicente", "Puerto marítimo de Coronel", "Puerto marítimo de Lirquen"]

start_port = "Puerto marítimo de Valparaiso"

# Ingresar los datos del barco y su carga
longitud = 165
ancho = 28
altura = 11
peso_barco = 30000
peso_carga = 40000
peso_carga_total = peso_barco + peso_carga
densidad = 1027

# Calcular el calado inicial del barco
calado_barco = calcular_calado(longitud, ancho, altura, peso_carga_total, densidad)

draft_limits = [calado_barco] * len(port_names)
peso_carga_actual = peso_carga

start_index = get_port_index(port_names, start_port)
if start_index != -1:
    path = nearest_neighbor(locations, port_names, start_port)
    total_distance = 0

    print("\nLa ruta más eficiente que cumple con los límites de calado y comienza en {} es:".format(start_port))

    for i in range(len(path)):
        port = path[i]
        if i == 0:
            print("\nPuerto:", port_names[port])
            print("Carga del barco que obtiene en el Puerto:", round(peso_carga_actual), "Tn")
            print("Calado del barco que obtiene en el Puerto: ", round(draft_limits[port], 3), "m")
            print("\n________________________________________\n")

        if i > 0:
            print("\nPuerto:", port_names[port])
            print("Carga del barco al llegar al puerto:", round(peso_carga_actual), "Tn")
            print("Calado del barco al llegar al puerto: ", round(draft_limits[port], 3), "m")
            print("Distancia recorrida hasta el momento: ", round(total_distance, 2), "KM")
            print("\n________________________________________\n")

        if port != start_index and i != len(path) - 1:
            peso_restante = float(input("Ingrese la cantidad de peso que se puede quitar en {}: ".format(port_names[port])))
            peso_carga_actual -= peso_restante
            calado_barco = calcular_calado(longitud, ancho, altura, peso_carga_actual, densidad)
            draft_limits[port] = calado_barco
            print("Carga del barco al salir del puerto:", round(peso_carga_actual), "Tn")
            print("Calado del barco al salir del puerto:", round(calado_barco, 3), "m")
            if peso_carga_actual <= 0:
              print("\nSe ha agotado la carga del barco. El barco regresará al puerto de origen:", start_port)
              total_distance = total_distance + locations[port_names.index(start_port)][port_names.index(port_names[port])]
              break

        if i < len(path) - 1:
            distance = locations[port][path[i+1]]
            total_distance = distance + total_distance
        else:
            total_distance = total_distance + locations[port_names.index(start_port)][port_names.index(port_names[port])]

    print("Distancia total recorrida: ", round(total_distance, 2), "KM")
else:
    print("El puerto de inicio no se encontró en la lista de puertos.")

def calculate_distance(locations, path):
    distance = 0
    for i in range(len(path) - 1):
        start = path[i]
        end = path[i + 1]
        distance += locations[start][end]
    return distance

def get_port_index(port_names, port_name):
    for i in range(len(port_names)):
        if port_names[i] == port_name:
            return i
    return -1

def get_nearest_neighbor(locations, unvisited_ports, current_port):
    min_distance = float('inf')
    nearest_port = None
    for port in unvisited_ports:
        if locations[current_port][port] < min_distance:
            min_distance = locations[current_port][port]
            nearest_port = port
    return nearest_port

def nearest_neighbor(locations, port_names, start_port):
    num_ports = len(locations)
    current_port = get_port_index(port_names, start_port)
    unvisited_ports = list(range(num_ports))
    unvisited_ports.remove(current_port)
    path = [current_port]

    while unvisited_ports:
        nearest_port = get_nearest_neighbor(locations, unvisited_ports, current_port)
        unvisited_ports.remove(nearest_port)
        path.append(nearest_port)
        current_port = nearest_port

    path.append(get_port_index(port_names, start_port))

    return path

# Función para calcular el calado del barco
def calcular_calado(longitud_barco, ancho_barco, altura_casco, peso_carga, densidad_agua):
    # Calcula el volumen del casco
    volumen_casco = longitud_barco * ancho_barco * altura_casco

    # Calcula el volumen de la carga
    volumen_carga = peso_carga / densidad_agua

    # Calcula el volumen total sumergido
    volumen_total = volumen_casco + volumen_carga

    # Calcula el área de la sección transversal del casco
    area_casco = longitud_barco * ancho_barco

    # Calcula el calado
    calado = volumen_total / area_casco

    return calado

# Datos de entrada
locations = [
    [0, 90, 402.03, 452.47, 404.75],
    [90, 0, 477.95, 514.14, 476.68],
    [402.03, 477.95, 0, 56.26, 53.72],
    [452.47, 514.14, 56.26, 0, 98.53],
    [404.75, 476.68, 53.72, 98.53, 0]
]
port_names = ["Puerto marítimo de San Antonio", "Puerto marítimo de Valparaiso", "Puerto marítimo de San Vicente", "Puerto marítimo de Coronel", "Puerto marítimo de Lirquen"]

start_port = "Puerto marítimo de Valparaiso"

# Ingresar los datos del barco y su carga
longitud = 165
ancho = 28
altura = 11
peso_barco = 30000
peso_carga = 40000
peso_carga_total = peso_barco + peso_carga
densidad = 1027

# Calcular el calado inicial del barco
calado_barco = calcular_calado(longitud, ancho, altura, peso_carga_total, densidad)

draft_limits = [calado_barco] * len(port_names)
peso_carga_actual = peso_carga

start_index = get_port_index(port_names, start_port)
if start_index != -1:
    path = nearest_neighbor(locations, port_names, start_port)
    total_distance = 0

    print("\nLa ruta más eficiente que cumple con los límites de calado y comienza en {} es:".format(start_port))

    for i in range(len(path)):
        port = path[i]
        if i == 0:
            print("\nPuerto:", port_names[port])
            print("Carga del barco que obtiene en el Puerto:", round(peso_carga_actual), "Tn")
            print("Calado del barco que obtiene en el Puerto: ", round(draft_limits[port], 3), "m")
            print("\n________________________________________\n")

        if i > 0:
            print("\nPuerto:", port_names[port])
            print("Carga del barco al llegar al puerto:", round(peso_carga_actual), "Tn")
            print("Calado del barco al llegar al puerto: ", round(draft_limits[port], 3), "m")
            print("Distancia recorrida hasta el momento: ", round(total_distance, 2), "KM")
            print("\n________________________________________\n")

        if port != start_index and i != len(path) - 1:
            peso_restante = float(input("Ingrese la cantidad de peso que se puede quitar en {}: ".format(port_names[port])))
            peso_carga_actual -= peso_restante
            calado_barco = calcular_calado(longitud, ancho, altura, peso_carga_actual, densidad)
            draft_limits[port] = calado_barco
            print("Carga del barco al salir del puerto:", round(peso_carga_actual), "Tn")
            print("Calado del barco al salir del puerto:", round(calado_barco, 3), "m")
            if peso_carga_actual <= 0:
              print("\nSe ha agotado la carga del barco. El barco regresará al puerto de origen:", start_port)
              total_distance = total_distance + locations[port_names.index(start_port)][port_names.index(port_names[port])]
              break

        if i < len(path) - 1:
            distance = locations[port][path[i+1]]
            total_distance = distance + total_distance
        else:
            total_distance = total_distance + locations[port_names.index(start_port)][port_names.index(port_names[port])]

    print("Distancia total recorrida: ", round(total_distance, 2), "KM")
else:
    print("El puerto de inicio no se encontró en la lista de puertos.")


fin = time.time()
print(fin-inicio)
# Obtener el consumo de memoria actual en bytes
memory_usage = resource.getrusage(resource.RUSAGE_SELF).ru_maxrss
print(f"Consumo de memoria: {memory_usage} bytes")
```
